## 2. Core Generation Function

```python
def generate(system_prompt: str, user_input: str, stream: bool = True) -> str:
    """
    Send a request to Claude and return the response text.
    If stream=True, prints output token-by-token as it arrives.
    """
    if stream:
        full_text = ""
        with client.messages.stream(
            model=MODEL,
            max_tokens=1024,
            system=system_prompt,
            messages=[{"role": "user", "content": user_input}],
        ) as stream_obj:
            for text in stream_obj.text_stream:
                print(text, end="", flush=True)
                full_text += text
        print()  # newline after streaming
        return full_text
    else:
        message = client.messages.create(
            model=MODEL,
            max_tokens=1024,
            system=system_prompt,
            messages=[{"role": "user", "content": user_input}],
        )
        return message.content[0].text

print("✅ generate() function defined")
```

## 🔄 Batch Mode — Run All Four Tools at Once

```python
def run_all(
    listing_details: str = SAMPLE_LISTING,
    lead_message: str = SAMPLE_LEAD,
    client_notes: str = SAMPLE_SUMMARY,
    listing_for_email: str = SAMPLE_EMAIL,
) -> dict:
    """
    Run all four RealtyAI tools sequentially.
    Returns a dict with keys: listing, lead, summary, email.
    """
    results = {}

    tools = [
        ("listing", "🏡 Listing Description", SYSTEM_LISTING, listing_details),
        ("lead",    "💬 Lead Reply",          SYSTEM_LEAD,    lead_message),
        ("summary", "📋 Client Summary",      SYSTEM_SUMMARY, client_notes),
        ("email",   "✉️  Marketing Email",     SYSTEM_EMAIL,   listing_for_email),
    ]

    for key, label, system, user_input in tools:
        print(f"\n{'='*60}")
        print(f"  {label}")
        print(f"{'='*60}")
        results[key] = generate(system, user_input, stream=True)

    print("\n✅ All four tools complete.")
    return results


# Run with sample data
all_results = run_all()
```

---
## 💾 Export Results to a Text File

```python
from datetime import datetime

def export_results(results: dict, filename: str = None) -> str:
    """Save all generated results to a .txt file."""
    if filename is None:
        ts = datetime.now().strftime("%Y%m%d_%H%M%S")
        filename = f"realtyai_results_{ts}.txt"

    sections = {
        "listing": "🏡 LISTING DESCRIPTION",
        "lead":    "💬 LEAD REPLY",
        "summary": "📋 CLIENT SUMMARY",
        "email":   "✉️  MARKETING EMAIL",
    }

    lines = ["RealtyAI — Generated Results", "=" * 60, ""]
    for key, label in sections.items():
        if key in results:
            lines += [label, "-" * 40, results[key], "", ""]

    with open(filename, "w", encoding="utf-8") as f:
        f.write("\n".join(lines))

    print(f"✅ Saved to: {filename}")
    return filename


# Export the batch results
export_results(all_results)
```
